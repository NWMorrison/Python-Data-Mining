# Sum Of Squared Differences
---

```python
import numpy as np


def sumOfSquaredDifferences(arr):
    N = len(arr)
    sum = 0

    # Nested loops to generate all possible pairs
    for i in range(N):
        for j in range(i + 1, N):
            # Calculating the square of the difference between the pairs
            diff = arr[i] - arr[j]
            sum += (diff * diff)

    # Output the sum of squared differences of all possible pairs
    print(sum)





x = [2,8,4]
sumOfSquaredDifferences(x)
```

```python
import numpy as np

# Define the arrays
C = np.array([
    [2, 5, 1],
    [4, 10, 2],
    [0, 5, 11],
    [1, 2, 100],
    [0, 15, 1]])

lbl = np.array([2, 1, 2, 1, 1])

# Create a boolean mask where lbl == 1
mask = lbl == 1

# Use the mask to filter C
result = C[mask]

print(result)

'''
1. import numpy as np
   This line imports the NumPy library and gives it the alias 'np',
   allowing us to use NumPy functions with the prefix 'np'.

2. C = np.array([...])` and `lbl = np.array([...])
   These lines create NumPy arrays from the given lists.
   C is a 5x3 array, and lbl is a 1D array with 5 elements.

3. mask = lbl == 1
   This line creates a boolean mask. Here's what happens:
   - NumPy compares each element of lbl to the value 1.
   - It returns a new array of the same shape as lbl,
      where each element is True if the corresponding
      element in lbl is 1, and False otherwise.
   - So, mask becomes: [False, True, False, True, True]

4. result = C[mask]
   This is where the filtering happens:
   - NumPy uses the boolean mask to index into C.
   - It selects only the rows of C where the corresponding element in mask is True.
   - In this case, it selects the 2nd, 4th, and 5th rows of C
   (remember, Python uses 0-based indexing).

5. print(result)
   This prints the resulting filtered array.

The key to this operation is NumPy's boolean indexing.
When you use a boolean array to index another array,
NumPy returns only the elements where the boolean array is True.
In this case, we're using it to select entire rows of C based
on the values in lbl.

This method is very efficient because:
1. It avoids explicit loops, using vectorized operations instead.
2. It doesn't require creating a new array manually;
    NumPy handles the creation of the new, filtered array.
3. It's concise and readable, expressing a complex operation
    in a single line of code.

The result is a new array containing only the rows of C where the corresponding
value in lbl is 1, achieving the filtering you requested without using explicit
loops or conditional statements.

'''
```

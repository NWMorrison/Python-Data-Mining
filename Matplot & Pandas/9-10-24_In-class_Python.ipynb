```python
df_iris = pd.read_csv('/content/iris.csv', header=None)

df_iris
```

```python

# Provides to rows & columns
print(df_iris.shape)

# Just provides the length of the Dataframe
print(len(df_iris))

# it is better to use loc or iloc, see documentation
df_iris.iloc[5:10]
```

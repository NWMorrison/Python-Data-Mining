```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/4.) Clustering Material/clustering_data.csv',header=None)

# the file has no column headers, so add headers
df.columns = ['x1', 'x2']
df.head(10)
```

```python
data = df.to_numpy()
```

```python
from sklearn.cluster import KMeans
import numpy as np

k=5 # change this to enter different number of clusters

# call k-means fit() to apply clustering on the data
# https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html
kmeans = KMeans(n_clusters=k, random_state=40, n_init="auto").fit(data)
```

```python
# results: cluster centroids, the centroid for each cluster
kmeans.cluster_centers_
```

```python
# cluster labels assigned to each point in the data
kmeans.labels_
```

```python
# clustering metrics
# there are many e.g. I am using silhouette score
# https://scikit-learn.org/stable/modules/clustering.html#clustering-performance-evaluation
from sklearn import metrics
from sklearn.metrics import pairwise_distances
metrics.silhouette_score(data, kmeans.labels_, metric='euclidean')
```

```python
# total SSE[Sum-Squared-Error], over ALL the clusters and data points, they call it inertia
kmeans.inertia_
```

```python
# SSE[Sum-Squared-Error] in EACH of the k clusters
# this is not code you are expected to know how to do on your own in this course
# just as an idea of how to find things (e.g. Google search or use any Gen AI tool)
# from https://stackoverflow.com/questions/66555997/how-to-get-sse-for-each-cluster-in-k-means
clusterwise_sse = np.zeros(k)
for point, label in zip(data, kmeans.labels_):
    clusterwise_sse[label] += np.square(point - kmeans.cluster_centers_[label]).sum()
```

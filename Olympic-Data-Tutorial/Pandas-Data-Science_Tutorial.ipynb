# **VII.) Adding Custom Columns To Our Dataframe**

## Bios

```python
# Putting individuals into a short, average, and tall category
# Uses a lambda function

bios['height_category'] = bios['height_cm'].apply(lambda x: 'Short' if x < 165 else('Average' if x < 185 else 'Tall'))

bios
```

```python
# Creating a weight category using lambda function
bios['weight'] = bios['weight_kg'].apply(lambda x: 'Lightweight' if x < 65 else('Middleweight' if (65 <= x <= 75) else 'Heavyweight'))

bios
```

```python
def categorize(row):
  if row['height_cm'] < 175 and row['weight_kg'] < 70:
    return 'Lightweight'

  elif row['height_cm'] < 185 and row['weight_kg'] < 80:
    return "Middleweight"

  else:
    return "Heavyweight"

bios['Category'] = bios.apply(categorize, axis=1)

bios
```

```python
new_df_coffee = coffee
new_df_coffee
```

```python
# Groupby() if we want to do certain calculations on certain data.
# .sum(), .mean()
new_df_coffee.groupby(['Coffee Type'])['Units Sold'].sum()
```

```python
# Inserting a dictionary to sum the units sold and then average the prices
new_df_coffee.groupby(['Coffee Type']).agg({'Units Sold' : 'sum', 'price' : 'mean'})
```

```python
# Group by multiple
# Could be useful to tell what sells the most on what days.

new_df_coffee.groupby(['Coffee Type' , 'Day']).agg({'Units Sold' : 'sum', 'price' : 'mean'})
```

## .Groupby()

```python
bios['born_date'] = pd.to_datetime(bios['born_date'])

bios.groupby(bios['born_date'].dt.year)['name'].count().reset_index().sort_values('name', ascending = False)
```

```python
# Let's say we want to group by the year born and month to see which months have
#       the most dominate athletes

bios['born_date'] = pd.to_datetime(bios['born_date'])

bios['month_born'] = bios['born_date'].dt.month
bios['year_born'] = bios['born_date'].dt.year


bios.groupby([bios['year_born'], bios['month_born']])['name'].count().reset_index().sort_values('name', ascending = False)
```

## .shift()



```python
# See how revenue is by-day or by-week for comparisons

adv_df_coffee['yesterday_revenue'] = coffee['revenue'].shift(2)

adv_df_coffee['revenue_difference'] = adv_df_coffee['revenue'] - adv_df_coffee['yesterday_revenue']

# Percentage of change by day
adv_df_coffee['%_of_change'] = ((adv_df_coffee['revenue'] / adv_df_coffee['yesterday_revenue']) * 100)


adv_df_coffee.info()



adv_df_coffee

```

## .rank()

```python
# Ranking athlete heights from tallest to shortest

bios['height_rank'] = bios['height_cm'].rank(ascending=False)
bios.sort_values(['height_rank'])

```

```python
# Sorting

bios.sort_values(['height_rank']).sample(10)[['name', 'height_rank']]
```

## .rolling()

```python
# using .rolling() to show revenue over 3 day periods

adv_df_coffee['3day_revenue'] = adv_df_coffee['revenue'].rolling(3).sum()

adv_df_coffee
```

## .cumsum()

```python
# using .cumsum() to calculate the cumlative sum up to that specific point in the row.

adv_df_coffee['cumulative_revenue'] = adv_df_coffee['revenue'].cumsum()

adv_df_coffee
```
